```python
import pandas as pd
```

```python
original_df = pd.read_csv('./data/anime.csv')
original_df.head()
```

```python
anime_names_df = pd.read_csv('./data/anime_filtered.csv')
anime_names_df = anime_names_df[['title', 'image_url']]
print(anime_names_df.shape)
anime_names_df.head()
```

```python
# Filter the DataFrame to keep only the rows in anime_names_df that have matching anime names in original_df
filtered_anime_names_df = anime_names_df.merge(original_df[['Name']], left_on='title', right_on='Name', how='inner')
filtered_anime_names_df = filtered_anime_names_df.drop(columns=['Name'])

# Remove rows where image url is missing
filtered_anime_names_df = filtered_anime_names_df.dropna(subset=['image_url'])

# Display the sorted DataFrame
print(filtered_anime_names_df.shape)
filtered_anime_names_df.head()
```

```python
# Remove the commas from the anime names
filtered_anime_names_df['title'] = filtered_anime_names_df['title'].str.replace(',', '')
```

```python
filtered_anime_names_df.to_csv('./data/anime_names.csv', sep=',', index=False)
```
